# ðŸ“¡ Simulating Sensor Data Streams
This project simulates real-time IoT sensor data, stores it, visualizes it, and saves it to a file.



```python
# Import standard Python libraries for data simulation and analysis
import random
import time
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
def generate_sensor_data():
    """
    Simulate readings from 3 sensors.

    Returns:
        dict: Dictionary containing timestamp and sensor readings
    """
    data = {
        'timestamp': datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
        'temperature_c': round(random.uniform(20.0, 35.0), 2),
        'humidity_percent': round(random.uniform(30.0, 90.0), 2),
        'light_lux': round(random.uniform(100, 1000), 2)
    }
    return data

# Test the function
print("Sample sensor reading:")
print(generate_sensor_data())
```

```python
print("Generating 5 sensor readings (1 per second):")
print("-" * 70)

for i in range(5):  # Generate 5 readings
    sensor_data = generate_sensor_data()
    print(f"Reading {i+1}: {sensor_data}")
    time.sleep(1)  # Wait 1 second between readings
```

```python
# Create an empty list to store data
data_stream = []

# Generate 20 sensor readings
print("Collecting 20 sensor readings...")
for i in range(20):
    reading = generate_sensor_data()
    data_stream.append(reading)
    time.sleep(0.2)  # Faster stream (200ms between readings)

# Convert to DataFrame for easier analysis
df = pd.DataFrame(data_stream)

print("\nDataFrame created!")
print(f"Shape: {df.shape} (rows, columns)")
print("\nFirst 5 rows:")
print(df.head())

print("\nDataFrame statistics:")
print(df.describe())
```

```python
# Create a figure with multiple subplots
fig, axes = plt.subplots(3, 1, figsize=(12, 8))

# Plot 1: Temperature over time
axes[0].plot(range(len(df)), df['temperature_c'], marker='o', color='red', label='Temperature')
axes[0].set_ylabel('Temperature (Â°C)', fontsize=11)
axes[0].set_title('Simulated Sensor Data Stream', fontsize=13, fontweight='bold')
axes[0].grid(True, alpha=0.3)
axes[0].legend()

# Plot 2: Humidity over time
axes[1].plot(range(len(df)), df['humidity_percent'], marker='s', color='blue', label='Humidity')
axes[1].set_ylabel('Humidity (%)', fontsize=11)
axes[1].grid(True, alpha=0.3)
axes[1].legend()

# Plot 3: Light intensity over time
axes[2].plot(range(len(df)), df['light_lux'], marker='^', color='orange', label='Light')
axes[2].set_xlabel('Reading Number', fontsize=11)
axes[2].set_ylabel('Light (lux)', fontsize=11)
axes[2].grid(True, alpha=0.3)
axes[2].legend()

plt.tight_layout()
plt.show()

print("âœ… Plots generated successfully!")
```

```python
# Save the DataFrame to a CSV file
filename = 'simulated_sensor_data.csv'
df.to_csv(filename, index=False)

print(f"âœ… Data saved to '{filename}'")
print(f"ðŸ“ Total records saved: {len(df)}")
print(f"ðŸ“‹ Columns: {', '.join(df.columns)}")
```

```python
# Calculate basic statistics
print("ðŸ“Š Data Analysis:")
print("=" * 50)

print(f"\nTemperature:")
print(f"  Min: {df['temperature_c'].min()}Â°C")
print(f"  Max: {df['temperature_c'].max()}Â°C")
print(f"  Mean: {df['temperature_c'].mean():.2f}Â°C")

print(f"\nHumidity:")
print(f"  Min: {df['humidity_percent'].min()}%")
print(f"  Max: {df['humidity_percent'].max()}%")
print(f"  Mean: {df['humidity_percent'].mean():.2f}%")

print(f"\nLight Intensity:")
print(f"  Min: {df['light_lux'].min()} lux")
print(f"  Max: {df['light_lux'].max()} lux")
print(f"  Mean: {df['light_lux'].mean():.2f} lux")
```

```python
# Example: Alert if temperature is too high or low
TEMP_MIN = 22
TEMP_MAX = 32

print("ðŸš¨ Temperature Alerts:")
print("=" * 50)

alerts = 0
for idx, row in df.iterrows():
    temp = row['temperature_c']
    if temp < TEMP_MIN:
        print(f"âš ï¸  Row {idx}: Temperature too LOW ({temp}Â°C)")
        alerts += 1
    elif temp > TEMP_MAX:
        print(f"âš ï¸  Row {idx}: Temperature too HIGH ({temp}Â°C)")
        alerts += 1

if alerts == 0:
    print("âœ… No temperature alerts - all readings within range!")
else:
    print(f"\nTotal alerts: {alerts}")
```
